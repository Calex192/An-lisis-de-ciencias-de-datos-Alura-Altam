```python
# Asegurar que 'Precio' es numérico
df_join['Precio'] = pd.to_numeric(df_join['Precio'], errors='coerce')

# Eliminar filas con valores nulos necesarios para el análisis
df_clean = df_join.dropna(subset=['Precio', 'Tienda', 'Categoría del Producto'])
ingresos_por_categoria = df_clean.groupby(['Tienda', 'Categoría del Producto'])['Precio'].sum().reset_index()
ingresos_por_categoria = ingresos_por_categoria.rename(columns={'Precio': 'Ingresos'})
ingresos_por_categoria.sort_values(by='Ingresos', ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.barplot(data=ingresos_por_categoria, x='Tienda', y='Ingresos', hue='Categoría del Producto')
plt.title('Ingresos por Tienda según Categoría de Producto')
plt.ylabel('Ingresos totales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**¡Extra! Análisis del desempeño geográfico**

En este apartado hare un análisis del desempeño de ingresos según la Categoría del Producto, además graficare los resultados para determinar las ventas realizadas de acuerdo a los datos geográficos proporcionados en el DataFrame.

```python

df_clean = df_join.dropna(subset=['Categoría del Producto', 'lat', 'lon', 'Tienda'])


df_clean['lat'] = pd.to_numeric(df_clean['lat'], errors='coerce')
df_clean['lon'] = pd.to_numeric(df_clean['lon'], errors='coerce')


df_clean = df_clean.dropna(subset=['lat', 'lon'])

ventas_categoria = df_clean.groupby(['Categoría del Producto', 'Tienda', 'lat', 'lon']) \
                           .size().reset_index(name='Ventas')
```

**Graficamos la distribución geográfica de las ventas según Categorías de Producto de cada Tienda.**

Distribución geográfica de "Electrodomésticos" en Tienda 1

```python
import folium
from folium.plugins import HeatMap

# Filtrar por categoría y tienda
cat = 'Electrodomésticos'
tienda = 'Tienda 1'

filtro = ventas_categoria[
    (ventas_categoria['Categoría del Producto'] == cat) &
    (ventas_categoria['Tienda'] == tienda)
]

# Crear mapa centrado
mapa = folium.Map(location=[filtro['lat'].mean(), filtro['lon'].mean()], zoom_start=5)

# Crear heatmap
heat_data = filtro[['lat', 'lon', 'Ventas']].values.tolist()
HeatMap(heat_data, radius=15).add_to(mapa)

mapa
```

**Determinar la ubicación geográfica de las tiendas.**

```python
# Crear mapa base
mapa = folium.Map(location=[ubicaciones_tiendas['lat'].mean(), ubicaciones_tiendas['lon'].mean()], zoom_start=5)

# Añadir cada tienda como marcador
for _, fila in ubicaciones_tiendas.iterrows():
    folium.Marker(
        location=[fila['lat'], fila['lon']],
        popup=f"{fila['Tienda']}",
        icon=folium.Icon(color='green', icon='shopping-cart', prefix='fa')
    ).add_to(mapa)

mapa
```
